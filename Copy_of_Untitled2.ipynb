```python
import pandas as pd

df = pd.read_csv("/content/SampleSuperstore.csv")

df.head()
```

```python
df.info()
```

```python
for col in df.columns:
    print(col)
```

```python
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
total_quantity = df['Quantity'].sum()

print("Total Sales =", total_sales)
print("Total Profit =", total_profit)
print("Total Quantity Sold =", total_quantity)
```

```python
import matplotlib.pyplot as plt

category_sales = df.groupby('Category')['Sales'].sum()

plt.figure(figsize=(6,6))
category_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title("Category-wise Sales")
plt.ylabel('')
plt.show()
```

```python
region_sales = df.groupby('Region')['Sales'].sum()

plt.figure(figsize=(8,5))
region_sales.plot(kind='bar')
plt.title("Region-wise Sales")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```

```python
subcat_sales = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(12,5))
subcat_sales.head(10).plot(kind='bar')
plt.title("Top 10 Sub-Categories by Sales")
plt.ylabel("Sales")
plt.show()
```

```python
profit_category = df.groupby('Category')['Profit'].sum()

plt.figure(figsize=(8,5))
profit_category.plot(kind='bar')
plt.title("Profit by Category")
plt.ylabel("Profit")
plt.show()
```

```python
state_sales = df.groupby('State')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(12,5))
state_sales.head(10).plot(kind='bar')
plt.title("Top 10 States by Sales")
plt.ylabel("Sales")
plt.show()
```

**INSIGHTS**

1. Technology category recorded the highest sales among all categories.
2. The West region contributed the highest overall revenue.
3. A few sub-categories generated a major share of total sales.
4. Sales performance varied significantly across different states.
5. Profit margins were not the same across all categories, indicating opportunities for optimization.

**RECOMMENDATIONS**

1. Increase marketing efforts for high-performing categories to maximize revenue.
2. Focus on improving sales in low-performing regions through targeted campaigns.
3. Maintain sufficient inventory for top-selling products and sub-categories.
4. Review discount strategies to improve overall profitability.
5. Expand successful sales strategies from high-performing states to other regions.
